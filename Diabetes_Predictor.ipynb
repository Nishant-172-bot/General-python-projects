Data Pre Processing

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler as StdSc
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python

df=pd.read_csv("/content/diabetes.csv")
df.shape
df.tail()
```

```python

df['Outcome'].value_counts()
```

```python
df.describe()
```

```python
df.groupby('Outcome').mean()
```

```python
X=df.drop(columns='Outcome',axis=1)
Y=df['Outcome']
print(X)
print(Y)
```

Data Standardization

```python
scale = StdSc()
scale.fit(X)
std_data=scale.transform(X)
#fit standard scaler onto a variable and tranforming data through that variable to a common range
```

```python
X=std_data
print(X)
```

Train Test Split

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.05,stratify=Y,random_state=2)
print(X_train.shape,X_test.shape)
```

Model Training

```python
classifier=svm.SVC(kernel='linear') #support vector classifier
classifier.fit(X_train,Y_train)
```

Model Evaluation

```python
#accuracy score of training data
X_Train_accuracy=classifier.predict(X_train)
training_accuracy=accuracy_score(X_Train_accuracy,Y_train)
print("Accuracy of Training: ",training_accuracy)
```

```python
#accuracy score of testing data
X_Test_accuracy=classifier.predict(X_test)
testing_accuracy=accuracy_score(X_Test_accuracy,Y_test)
print("Accuracy of Testing: ",testing_accuracy)
```

Predictor System

```python
#taking input dtat converting to numpy array and reshape to indicate 1 data
Input_data=(3,126,88,41,235,39.3,0.704,27)
input_as_nympy_array=np.asarray(Input_data)
input_data_reshaped=input_as_nympy_array.reshape(1,-1)

#standardizing as per scaler
std_data=scale.transform(input_data_reshaped)
print(std_data)
prediction=classifier.predict(std_data)
if prediction[0]==0:
  print("Non Diabetic")
else:
  print("Diabetic")
```
